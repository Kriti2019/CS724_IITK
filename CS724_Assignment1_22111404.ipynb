```python
import numpy as np
import math
import random
#Given user's location 
loc=(100,100,100)

#select position of satellite 1
satellite_x1=5600
satellite_y1=400
satellite_z1=600
s1=(satellite_x1,satellite_y1,satellite_z1)

#select position of satellite 2
satellite_x2=700
satellite_y2=800
satellite_z2=300
s2=(satellite_x2,satellite_y2,satellite_z2)

#select position of satellite 3
satellite_x3=500
satellite_y3=800
satellite_z3=400
s3=(satellite_x3,satellite_y3,satellite_z3)

#select position of satellite 4
satellite_x4=300
satellite_y4=700
satellite_z4=400
s4=(satellite_x4,satellite_y4,satellite_z4)

#select position of satellite 5
satellite_x5=144
satellite_y5=218
satellite_z5=370
s5=(satellite_x5,satellite_y5,satellite_z5)

#finding distance of satellites from user
d1=math.dist(loc,s1)
d2=math.dist(loc,s2)
d3=math.dist(loc,s3)
d4=math.dist(loc,s4)
d5=math.dist(loc,s5)

#speed of light
c=300000000

t1=d1/c
t2=d2/c
t3=d3/c
t4=d4/c
t5=d5/c
timeMatrix=np.array([t1,t2,t3,t4,t5])

#Displaying the time
print("Time to reach the signal from the object from Satellite 1 is", t1,"seconds")
print("Time to reach the signal from the object from Satellite 2 is", t2,"seconds")
print("Time to reach the signal from the object from Satellite 3 is", t3,"seconds")
print("Time to reach the signal from the object from Satellite 4 is", t4,"seconds")
print("Time to reach the signal from the object from Satellite 5 is", t5,"seconds")
```

```python
#Calculation Location of User By reverse Process
M = np.array ([[2*(satellite_x2-satellite_x1), 2*(satellite_y2-satellite_y1), 2*(satellite_z2-satellite_z1)],
               [2*(satellite_x3-satellite_x2), 2*(satellite_y3-satellite_y2), 2*(satellite_z3-satellite_z2)], 
               [2*(satellite_x4-satellite_x3), 2*(satellite_y4-satellite_y3), 2*(satellite_z4-satellite_z3)] 
              ])
N = np.array([[(d1**2-d2**2)-(satellite_x1**2-satellite_x2**2)-(satellite_y1**2-satellite_y2**2)-(satellite_z1**2-satellite_z2**2)],
             [(d2**2-d3**2)-(satellite_x2**2-satellite_x3**2)-(satellite_y2**2-satellite_y3**2)-(satellite_z2**2-satellite_z3**2)],
             [(d3**2-d4**2)-(satellite_x3**2-satellite_x4**2)-(satellite_y3**2-satellite_y4**2)-(satellite_z3**2-satellite_z4**2)]
          ])
X=np.linalg.solve(M,N)

print("The Estimated location of the user by reverse process is (",X[0],X[1],X[2],")")

```

We can See the Estimated location of the user by Opposite Operation is exactly same as the actual location of the user.
